```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow for Colab compatibility


# Load the image in grayscale
image = cv2.imread('/content/field-6574455_640.jpg', cv2.IMREAD_GRAYSCALE)

# Apply global histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display original and equalized images using cv2_imshow
cv2_imshow(image) # Use cv2_imshow instead of cv2.imshow
print()
cv2_imshow(equalized_image) # Use cv2_imshow instead of cv2.imshow


# Save the equalized image
cv2.imwrite('global_hist_equalized.jpg', equalized_image)

# cv2.waitKey(0) # These lines are not needed when using cv2_imshow in Colab
# cv2.destroyAllWindows() # These lines are not needed when using cv2_imshow in Colab
```
